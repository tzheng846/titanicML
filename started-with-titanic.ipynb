```python
train_data = pd.read_csv("/kaggle/input/titanic/train.csv")
train_data.sort_values("Fare",ascending=False).head(10)
```

```python
women = train_data.loc[train_data.Sex == 'female']["Survived"]
print("women", sum(women)/len(women))
men = train_data.loc[train_data.Sex == "male"]["Survived"]
print("men", sum(men)/len(men))
child = train_data.loc[train_data.Age <=15]["Survived"]
print("child", sum(child)/len(child))
adults = train_data.loc[train_data.Age >15]["Survived"]
print("adults", sum(adults)/len(adults))
rich = train_data.loc[train_data.Fare >train_data.Fare.quantile(0.80)]["Survived"]
print("rich", sum(rich)/len(rich))
poor = train_data.loc[train_data.Fare <train_data.Fare.quantile(0.80)]["Survived"]
print("poor", sum(poor)/len(poor))
```
